```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from skimage import exposure

# Load the image using the PIL library
image1 = Image.open('low_contrast3.jpg')

# Convert the image to a NumPy array
image_array1 = np.array(image1)

# Apply CLAHE for contrast enhancement
# Contrast Limited Adaptive Histogram Equalization (CLAHE)
clahe1 = exposure.equalize_adapthist(image_array1, clip_limit=0.03) # You can adjust the clip_limit parameter

# Convert the NumPy array back to an image
enhanced_image1 = Image.fromarray((clahe1 * 255).astype(np.uint8))

# Save the enhanced image
enhanced_image1.save('enhanced_image_clahe1.jpg')

# Display the original and enhanced images using Matplotlib
plt.subplot(121), plt.imshow(image_array1, cmap='gray')
plt.title('Original Image')
plt.subplot(122), plt.imshow(clahe1, cmap='gray')
plt.title('Enhanced Image (CLAHE)')
plt.show()
```
